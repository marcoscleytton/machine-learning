Observação sobre a atividade

Esta atividade consistiu em simular o resultado de um sistema de Recuperação de Informação (RI).
Foram criadas variáveis fixas para representar três componentes:

tokens: lista de palavras-chave extraídas de um documento ou consulta.

distância de Levenshtein: valor inteiro simulando a diferença entre strings.

similaridade Bag-of-Words: valor em ponto flutuante simulando o grau de semelhança entre textos.

Trata-se de uma atividade de simulação, pois os valores são atribuídos diretamente e não resultam de cálculos reais. O foco é demonstrar compreensão dos conceitos de RI e praticar a criação de variáveis em Python de forma organizada.

```python
def simular_ri():
    """
    Simula o resultado de um sistema de Recuperação de Informação (RI).

    Retorna três componentes fixos que representam:
    - tokens: lista de palavras-chave extraídas de uma consulta ou documento.
    - distancia_levenshtein: número inteiro simulando a diferença entre consulta e documento.
    - similaridade_bow: valor de similaridade Bag-of-Words (0 a 1).
    """
    # Lista de tokens que representam as palavras-chave extraídas
    tokens = ["busca", "informação", "texto"]

    # Distância de Levenshtein simulando diferença entre consulta e documento
    distancia_levenshtein = 2

    # Similaridade Bag-of-Words indicando grau de semelhança (0 = nenhuma, 1 = idêntico)
    similaridade_bow = 0.85

    return tokens, distancia_levenshtein, similaridade_bow
```
