Observao sobre a atividade

Esta atividade consistiu em simular o resultado de um sistema de Recuperao de Informao (RI).
Foram criadas variveis fixas para representar trs componentes:

tokens: lista de palavras-chave extradas de um documento ou consulta.

distncia de Levenshtein: valor inteiro simulando a diferena entre strings.

similaridade Bag-of-Words: valor em ponto flutuante simulando o grau de semelhana entre textos.

Trata-se de uma atividade de simulao, pois os valores so atribudos diretamente e no resultam de clculos reais. O foco  demonstrar compreenso dos conceitos de RI e praticar a criao de variveis em Python de forma organizada.

```python
def simular_ri():
    """
    Simula o resultado de um sistema de Recuperao de Informao (RI).

    Retorna trs componentes fixos que representam:
    - tokens: lista de palavras-chave extradas de uma consulta ou documento.
    - distancia_levenshtein: nmero inteiro simulando a diferena entre consulta e documento.
    - similaridade_bow: valor de similaridade Bag-of-Words (0 a 1).
    """
    # Lista de tokens que representam as palavras-chave extradas
    tokens = ["busca", "informao", "texto"]

    # Distncia de Levenshtein simulando diferena entre consulta e documento
    distancia_levenshtein = 2

    # Similaridade Bag-of-Words indicando grau de semelhana (0 = nenhuma, 1 = idntico)
    similaridade_bow = 0.85

    return tokens, distancia_levenshtein, similaridade_bow
```
